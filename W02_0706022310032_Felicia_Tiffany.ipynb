```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
# Handle missing values
# electricity

df["electricity"] = (
    df["electricity"]
    .astype(str)
    .str.replace("VA", "", regex=False)
    .str.replace(r"[^0-9]", "", regex=True)
)

df["electricity"] = pd.to_numeric(df["electricity"], errors="coerce")

print(df["electricity"].unique())
```

```python
# Handle missing values
# electricity

mask_missing = df["electricity"].isna()

median_elec = df["electricity"].median()
df["electricity"] = df["electricity"].fillna(median_elec)

print(df.loc[mask_missing, ["electricity", "building_size_m2", "bedrooms", "bathrooms", "land_size_m2"]])
```

```python
# Handle missing values
# property_type use a measure of central tendency for the attribute
# memilih modus untuk mengisi missing values karena hanya ada "rumah" dan nan, serta hanya ada 1 missing values

df['property_type'].fillna(df['property_type'].mode()[0], inplace=True)
```

```python
# Handle missing values
# certificate use a global constant to fill in the missing value
# karena jenis sertifikat tidak bisa di manipulasi jadi lebih baik untuk mengisinya dengan unknown

df['certificate'] = df['certificate'].fillna("unknown")
```

```python
# Handle missing values
# year_built di drop
# karena memberikan informasi yang sama dengan column building_age

df = df.drop(columns=["year_built"])
```

```python
# Handle missing values
# url dan ads_id di drop
# karena tidak memberikan informasi untuk visualisasi data

df = df.drop(columns=["url", "ads_id"])
```

```python
# Handle missing values
# menghitung jumlah fasilitas dengan split ',' dan menggunakan "set" agar tidak ada data duplikat dihitung 2 kali

df["facilities"] = df["facilities"].apply(
    lambda x: len(set(str(x).split(","))) if pd.notna(x) else 0
)
```

**Data Binning**

```python
# data binning untuk facilities
# agar informasi lebih mudah dibaca dengan dihitung jumlah fasilitas lalu dikelompokkan dalam data binning

df["facilities_bin"] = 0

df.loc[(df['facilities'] >= 1) & (df['facilities'] <= 5), 'facilities_bin'] = 1
df.loc[(df['facilities'] >= 6) & (df['facilities'] <= 10), 'facilities_bin'] = 2
df.loc[(df['facilities'] >= 11) & (df['facilities'] <= 15), 'facilities_bin'] = 3
df.loc[(df['facilities'] >= 16) & (df['facilities'] <= 20), 'facilities_bin'] = 4
df.loc[df['facilities'] >= 21, 'facilities_bin'] = 5

print(df[["facilities", "facilities_bin"]].head(10))
print("Unique bins:", df["facilities_bin"].unique())
```

```python
# Handle missing values
# building_orientation di drop
# karena jumlah missing values hampir 50% dantidak bisa dimanipulasi jadi lebih baik di drop

df = df.drop(columns=["building_orientation"])
```

```python
# Handle missing values
# floors use a measure of central tendency for the attribute
# menggunakan median karena datanya akan aman karena median mengambil nilai tengah

mask_missing = df["floors"].isna()

# hitung median
median_floors = df["floors"].median()
df["floors"] = df["floors"].fillna(median_floors)

print(df.loc[mask_missing, ["floors", "building_size_m2", "bedrooms", "bathrooms", "land_size_m2"]])
```

```python
# Handle missing values
# Deleting row untuk data yang missing value lebih dari 6
# data tersebut missing pada kolom yang penting seperti: bedroom, bathroom, land size dan building size

# Hitung missing value per row
df['missing_count'] = df.isna().sum(axis=1)

# Drop baris dengan missing value >= 6
df = df[df['missing_count'] < 6]

# Hapus kolom bantu
df = df.drop(columns=['missing_count'])
```

```python
# Handle missing values
# bedrooms use the most probable value to fill in the missing value
# karena bedrooms memiliki hubungan yang kuat antara building size, land size dan floors
from sklearn.tree import DecisionTreeRegressor
import numpy as np

features_bed = ["building_size_m2", "land_size_m2", "floors"]

mask_bed_missing = df["bedrooms"].isna()

# Train & Test
train_bed = df[df["bedrooms"].notna()]
test_bed = df[df["bedrooms"].isna()]

# Latih model & prediksi
if not test_bed.empty:
    model_bed = DecisionTreeRegressor(random_state=42)
    model_bed.fit(train_bed[features_bed], train_bed["bedrooms"])
    preds = model_bed.predict(test_bed[features_bed])
    df.loc[mask_bed_missing, "bedrooms"] = np.round(preds).astype(int)

print(df.loc[mask_bed_missing, ["bedrooms", "building_size_m2", "land_size_m2", "floors"]])
```

```python
# Handle missing values
# bathrooms use the most probable value to fill in the missing value
# karena bathrooms memiliki hubungan yang kuat antara bedrooms, building size, land size dan floors

from sklearn.tree import DecisionTreeRegressor
features_bath = ["building_size_m2", "land_size_m2", "floors", "bedrooms"]

mask_bath_missing = df["bathrooms"].isna()

# Train & Test
train_bath = df[df["bathrooms"].notna()]
test_bath = df[df["bathrooms"].isna()]

if not test_bath.empty:
    model_bath = DecisionTreeRegressor(random_state=42)
    model_bath.fit(train_bath[features_bath], train_bath["bathrooms"])
    preds = model_bath.predict(test_bath[features_bath])
    df.loc[mask_bath_missing, "bathrooms"] = np.round(preds).astype(int)

print(df.loc[mask_bath_missing, ["bathrooms", "building_size_m2", "land_size_m2", "floors", "bedrooms"]])
```

```python
# Handle missing values
# building_age use a measure of central tendency for the attribute
# paling cocok untuk mengisi building_age karena nilai tengah
mask_missing = df["building_age"].isna()

# hitung median
median_age = df["building_age"].median()
df["building_age"] = df["building_age"].fillna(median_age)

print(df.loc[mask_missing, ["building_age", "property_condition", "furnishing"]])
```

```python
# Handle missing values
# property_condition dan furnishing
# cara pertama fill the missing value manually, karena dilihat dari unique, terlihat ada value yang tertukar
# seharusnya "semi furnished" dann "unfurnished" terdapat di kolom furnishing
# dan "baru" terdapat di kolom property_condition, jadi saya menukar value dulu sebelum mengisi missing value yang lain

mask = df["property_condition"].isin(["semi furnished", "unfurnished"])

# simpan nilai yang salah
temp = df.loc[mask, "property_condition"]

# tukar: isi property_condition dengan furnishing
df.loc[mask, "property_condition"] = df.loc[mask, "furnishing"]

# isi furnishing dengan nilai yang tadinya salah
df.loc[mask, "furnishing"] = temp
```

```python
df["property_condition"].isna().sum()
```

```python
# Handle missing values
# property_condition use a measure of central tendency for the attribute
# memilih modus karena datanya kategorical

df['property_condition'].fillna(df['property_condition'].mode()[0], inplace=True)
```

```python
df["furnishing"].isna().sum()
```

```python
# Handle missing values
# furnishing use a measure of central tendency for the attribute
# memilih modus karena datanya kategorical

df['furnishing'].fillna(df['furnishing'].mode()[0], inplace=True)
```

```python
# binning price_in_rp
# karena banyak sekali unique value jadi lebih baik dikelompokkan ke dalam range tertentu
# saya memilih kategori ini karena harga kelompok yang realistis dan mudah dipahami audiens
# kelompok ini biasa digunakan oleh developer, agen dan bank

df["price_bin"] = 0

bins = [0, 5e8, 1e9, 5e9, 2e10, np.inf]
labels = ["<500 Jt", "500 Jtâ€“1 M", "1â€“5 M", "5â€“20 M", ">20 M"]

df["price_bin"] = pd.cut(df["price_in_rp"], bins=bins, labels=labels, right=True)

print(df[["price_in_rp", "price_bin"]].head(10))
```
